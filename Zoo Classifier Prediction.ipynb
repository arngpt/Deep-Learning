Data Augmentation

```python
from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=60,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,)

# Note that the validation data should not be augmented!
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        # This is the target directory
        train_dir,
        # All images will be resized to 150x150
        target_size=(150, 150),
        batch_size=20,
        # Since we use binary_crossentropy loss, we need binary labels
        class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(
        validation_dir,
        target_size=(150, 150),
        batch_size=20,
        class_mode='categorical')
```

```python
train_generator.class_indices
```

```python
for data_batch, labels_batch in train_generator:
    print('data batch shape:', data_batch.shape)
    print('labels batch shape:', labels_batch.shape)
    break
```

Creating Model

```python
from keras import layers
from keras import models

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                        input_shape=(150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dropout(0.2))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(3, activation='softmax'))
```

```python
model.summary()
```

```python
from tensorflow import keras
from keras import optimizers

model.compile(loss='binary_crossentropy',
              optimizer=keras.optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])
```

Running the model

```python
history = model.fit_generator(
      train_generator,
      steps_per_epoch=700//20,
      epochs=50,
      validation_data=validation_generator,
      validation_steps=150//20)
```

```python
validation_generator.class_indices
```

Saving model

Loading the model

```python
model = keras.models.load_model('cats_dogs_panda_save.h5')
```

Finding accuracy on training set

```python
test_generator = test_datagen.flow_from_directory(
test_dir,
target_size=(150, 150),
batch_size=20,
class_mode='categorical')



test_loss, test_acc = model.evaluate(test_generator)
print('Test accuracy:', test_acc)
print('Test Loss:', test_loss)
```

Loading a pretrained Model VGG19

```python
conv_base= VGG19(input_shape=(150,150,3),include_top=False, weights='imagenet')
```

```python
conv_base.summary()
```

Adding Layers to the model

```python
for layer in conv_base.layers:
    layer.trainable=False
```

```python


model1 = models.Sequential()
model1.add(conv_base)
model1.add(layers.Flatten())
model1.add(layers.Dropout(0.2))
model1.add(layers.Dense(512, activation='relu'))
model1.add(layers.Dense(3, activation='softmax'))
```

```python
model1.summary()
```

```python
from tensorflow import keras
from keras import optimizers

model1.compile(loss='binary_crossentropy',
              optimizer=keras.optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])
```
